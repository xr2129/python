做为 Apple Store App 独立开发者，你要搞限时促销，为你的应用生成激活码（或者优惠券），使用 Python 如何生成 200 个激活）？  
Create 200 coupon code for your app.

```python
import uuid
from dataclasses import dataclass
```

```python
@dataclass
class Coupon:
    """prototype for Coupon"""
    code: str = None
    item_id: str = None
        
```

```python
@dataclass
class CouponGen:
    '''
    Generate and return coupons
    '''
    def generateCoupon(sef,q, item):
        _list =[]
        
        for i in range(q):
            c = Coupon()
            c.code = str(uuid.uuid4()).upper()
            c.item_id = item
            _list.append(c)
            
        return _list
```

```python
if __name__ == '__main__':
    c = CouponGen()
    cl1 = c.generateCoupon(200,'0001')
    
    print(len(cl1))
    print('-----'*20)
    for i in cl1:
        print(i)
```
