```python
import PIL.Image
from PIL import ImageDraw
from tkinter import filedialog
from tkinter import Tk
from matplotlib.pyplot import imshow
import numpy as np
```

```python
def showImg(file):
    with PIL.Image.open(file) as img:
        %matplotlib inline
        imshow(np.asarray(img))
```

```python
def drawDot(file):
    with PIL.Image.open(file) as img:
        draw = ImageDraw.Draw(img)
        width, height = img.size
        draw.ellipse((width-width/8, 0, width, width/8), fill=(255, 0, 0))
        img.save('dotimg.png')
```

```python
if __name__ == '__main__':
    root = Tk()
    root.fpath = filedialog.askopenfilename()
    drawDot(root.fpath)
    showImg('dotimg.png')
    exit()
```
